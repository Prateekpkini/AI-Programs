```python
from collections import deque

class Graph:
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list.get(v, [])  # Use .get() to handle missing keys

    def h(self, n):
        H = {
            'M': 5,
            'N': 10,
            'O': 15,
            'P': 6,
            'Q': 7,
            'R': 17
        }
        return H.get(n, float('inf'))  # Default to infinity if node not found

    def a_star_algorithm(self, start_node, stop_node):
        if start_node not in self.adjacency_list or stop_node not in self.adjacency_list:
            print('Start or stop node not in graph!')
            return None

        open_list = set([start_node])
        closed_list = set([])

        g = {}
        g[start_node] = 0

        parents = {}
        parents[start_node] = start_node

        while len(open_list) > 0:
            n = None

            for v in open_list:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n is None:
                print('Path does not exist!')
                return None

            if n == stop_node:
                reconst_path = []

                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]

                reconst_path.append(start_node)
                reconst_path.reverse()

                print('Path found: {}'.format(reconst_path))
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n

                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            open_list.remove(n)
            closed_list.add(n)

        print('Path does not exist!')
        return None


# Example usage
adjacency_list = {
    'M': [('N', 5), ('O', 6), ('P', 13)],
    'N': [('O', 7), ('P', 2)],  # Add a connection from N to P
    'O': [('R', 12)],
    'P': [('Q', 10)],
    'Q': [('R', 4)]
}

graph1 = Graph(adjacency_list)
graph1.a_star_algorithm('N', 'Q')
```
